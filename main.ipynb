```python
import plotly.figure_factory as plotly_figure_factory
import plotly.graph_objects as plotly_graph_objects
import statistics as stats
import pandas as pd
import random as rand
import csv

# reading csv file
data_read = pd.read_csv("data.csv")
data_to_list = data_read["Math_score"].tolist()


# finding mean and standard deviation of the data_to_list
mean = stats.mean(data_to_list)
std = stats.stdev(data_to_list)

# printing the mean and standard deviation
print("mean: ", mean)
print("standard deviation: ", std)

# ploting graph
figure = plotly_figure_factory.create_distplot(
        [data_to_list], ["Math_score"], show_hist=False)
figure.show()

def generate_sample_data(counter):
    # generating sample data
    sample_data = []
    for i in range(0, counter):
        random_index = rand.randint(0, len(data_to_list) - 1)
        value = data_to_list[random_index]
        sample_data.append(value)
    
    mean = stats.mean(sample_data)

    return mean

hundred_val = []
for i in range(0, 1000):
    set_mean_value = generate_sample_data(100)
    hundred_val.append(set_mean_value)

#finding the mean and standard deviation of the hundred_val
hundred_mean = stats.mean(hundred_val)
hundred_std = stats.stdev(hundred_val)

print("mean of hundred_val",hundred_mean)
print("std of hundred_val",hundred_std)

#ploting the mean of hundred_val
figure2 = plotly_figure_factory.create_distplot(
        [hundred_val], ["SampleData"], show_hist=False)

figure2.add_trace(plotly_graph_objects.Scatter(x=[mean,hundred_mean], y=[0,0.2],mode="lines",name="mean"))
figure2.show()

# calculating the start standard deviation and end standard deviation
first_std_deviation_start, first_std_deviation_end = hundred_mean-hundred_std, hundred_mean+hundred_std
second_std_deviation_start, second_std_deviation_end = hundred_mean-(2*hundred_std), hundred_mean+(2*hundred_std)
third_std_deviation_start, third_std_deviation_end = hundred_mean-(3*hundred_std), hundred_mean+(3*hundred_std)

# printing all the first_std_deviation_start, first_std_deviation_end, second_std_deviation_start, second_std_deviation_end, third_std_deviation_start, third_std_deviation_end
print("std1",first_std_deviation_start, first_std_deviation_end)
print("std2",second_std_deviation_start, second_std_deviation_end)
print("std3",third_std_deviation_start,third_std_deviation_end)


#reading csv file data_read_1, data_read_2, data_read_3
data_read_1 = pd.read_csv("data_new_1.csv")
data_read_2 = pd.read_csv("data_new_2.csv")
data_read_3 = pd.read_csv("data_new_3.csv")

# converting it into list 
data_read_1_list = data_read_1["Math_score"].tolist()
data_read_2_list = data_read_2["Math_score"].tolist()
data_read_3_list = data_read_3["Math_score"].tolist()

# finding the mean of the three files
mean_data_read_1 = stats.mean(data_read_1_list)
mean_data_read_2 = stats.mean(data_read_2_list)
mean_data_read_3 = stats.mean(data_read_3_list)

#blank print
print("")

#printing mean of the three files
print("mean of data_read_1: ", mean_data_read_1)
print("mean of data_read_2: ", mean_data_read_2)
print("mean of data_read_3: ", mean_data_read_3)

print("")
#printing 1st heading
print("THE STUDENTS WHO GAVE EXTRA TIME TO MATH LAB")

# THE STUDENTS WHO GAVE EXTRA TIME TO MATH LAB and plotting on graph
fig = plotly_figure_factory.create_distplot([hundred_val],["student marks"], show_hist=False)
fig.add_trace(plotly_graph_objects.Scatter(x=[hundred_mean,hundred_mean], y=[0,0.17],mode="lines",name="mean"))
fig.add_trace(plotly_graph_objects.Scatter(x=[mean_data_read_1, mean_data_read_1], y=[0,0.17],mode="lines",name="MEAN OF STUDENTS WHO HAD MATH LABS"))
fig.add_trace(plotly_graph_objects.Scatter(x=[first_std_deviation_start, first_std_deviation_start], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 1 START"))
fig.add_trace(plotly_graph_objects.Scatter(x=[first_std_deviation_end, first_std_deviation_end], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 1 END"))
fig.add_trace(plotly_graph_objects.Scatter(x=[second_std_deviation_start, second_std_deviation_start], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 2 START"))
fig.add_trace(plotly_graph_objects.Scatter(x=[second_std_deviation_end, second_std_deviation_end], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 2 END"))
fig.add_trace(plotly_graph_objects.Scatter(x=[third_std_deviation_start, third_std_deviation_start], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 3 START"))
fig.add_trace(plotly_graph_objects.Scatter(x=[third_std_deviation_end, third_std_deviation_end], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 3 END"))
fig.show()

# printing 2nd heading
print("THE STUDENTS WHO USED MATH PRACTISE APP")

# finding the mean of the STUDENTS WHO USED MATH PRACTISE APP and plotting it on the plot.
fig2 = plotly_figure_factory.create_distplot([hundred_val],["student marks"], show_hist=False)
fig2.add_trace(plotly_graph_objects.Scatter(x=[hundred_mean,hundred_mean], y=[0,0.17],mode="lines",name="mean"))
fig2.add_trace(plotly_graph_objects.Scatter(x=[mean_data_read_2, mean_data_read_2], y=[0,0.17],mode="lines",name="MEAN OF STUDENTS WHO HAD MATH PRACTISE APP"))
fig2.add_trace(plotly_graph_objects.Scatter(x=[first_std_deviation_start, first_std_deviation_start], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 1 START"))
fig2.add_trace(plotly_graph_objects.Scatter(x=[first_std_deviation_end, first_std_deviation_end], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 1 END"))
fig2.add_trace(plotly_graph_objects.Scatter(x=[second_std_deviation_start, second_std_deviation_start], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 2 START"))
fig2.add_trace(plotly_graph_objects.Scatter(x=[second_std_deviation_end, second_std_deviation_end], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 2 END"))
fig2.add_trace(plotly_graph_objects.Scatter(x=[third_std_deviation_start, third_std_deviation_start], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 3 START"))
fig2.add_trace(plotly_graph_objects.Scatter(x=[third_std_deviation_end, third_std_deviation_end], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 3 END"))
fig2.show()


# printing 3rd heading
print("THE STUDENTS WHO WERE ENFORCED WITH REGISTERS")

# the STUDENTS WHO WERE ENFORCED WITH REGISTERS and plotting it on the plot.
fig3 = plotly_figure_factory.create_distplot([hundred_val],["student marks"], show_hist=False)
fig3.add_trace(plotly_graph_objects.Scatter(x=[hundred_mean,hundred_mean], y=[0,0.17],mode="lines",name="mean"))
fig3.add_trace(plotly_graph_objects.Scatter(x=[mean_data_read_3, mean_data_read_3], y=[0,0.17],mode="lines",name="MEAN OF STUDENTS WHO WERE ENFORCED WITH REGISTERS"))
fig3.add_trace(plotly_graph_objects.Scatter(x=[first_std_deviation_start, first_std_deviation_start], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 1 START"))
fig3.add_trace(plotly_graph_objects.Scatter(x=[first_std_deviation_end, first_std_deviation_end], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 1 END"))
fig3.add_trace(plotly_graph_objects.Scatter(x=[second_std_deviation_start, second_std_deviation_start], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 2 START"))
fig3.add_trace(plotly_graph_objects.Scatter(x=[second_std_deviation_end, second_std_deviation_end], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 2 END"))
fig3.add_trace(plotly_graph_objects.Scatter(x=[third_std_deviation_start, third_std_deviation_start], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 3 START"))
fig3.add_trace(plotly_graph_objects.Scatter(x=[third_std_deviation_end, third_std_deviation_end], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 3 END"))
fig3.show()


#finding the z score using the formula
z_score = (hundred_mean - mean_data_read_2)/hundred_std
print("Z SCORE IS:",z_score)
```
